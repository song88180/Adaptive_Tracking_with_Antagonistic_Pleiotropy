```python
import pandas as pd
import re
```

```python
df_all = pd.read_csv('FB_names_negative.csv',header=None,names=['marker gene'])
```

```python
df_all
```

```python
df_all = df_all[df_all['marker gene'].notna()]
```

```python
#df_all['marker gene'].to_csv('FB_names.csv',index=False)
```

```python
flag = False
for FB_name in df_all['marker gene']:
    output = ''
    N = 0
    with open('dmel-all-CDS-r6.46.fasta','r') as f:
        while True:
            line = f.readline()
            if not line: 
                break
            if (line[0] == '>'): 
                flag = False
                if N == 1:
                    break
                elif FB_name in line:
                    N += 1
                    flag = True
                    output += line
                    #result = re.search('>(.*) ', line)
                    #print(result.group(1))
            elif flag:
                output += line
    #print(output)
    with open(f'NC_genes/{FB_name}.fasta','w') as fw:
        fw.write(output)
    print(FB_name,f': {N} form')
```
