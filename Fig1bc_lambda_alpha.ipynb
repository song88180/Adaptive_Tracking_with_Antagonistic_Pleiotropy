```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import pickle as pkl
```

```python
with open('alpha_Ne_dict.pkl','rb') as f:
    alpha_Ne_dict = pkl.load(f)
with open('lambda_Ne_dict.pkl','rb') as f:
    lambda_Ne_dict = pkl.load(f)
```

```python
rcParams_dict = {
    'font.size': 15,
    'pdf.fonttype':42,
    'font.family':'Arial',
    'mathtext.fontset':'custom',
    'mathtext.default': 'regular',
    'mathtext.it': 'Arial:italic',
    'mathtext.it': 'Arial:italic',
}
mpl.rc('text', usetex = False)
mpl.rcParams.update(rcParams_dict)
```

```python
lambda_Ne_dict['1e+07']['DHFR'].keys()
```

```python
key_pair_list = \
    [('Yeast21genes','YPD'),('HSP90','standard_rep1'),('Ubiquitin','standard'),('DHFR','no_Lon')]
```

```python
lambda_dict = lambda_Ne_dict['1e+07']
alpha_dict = alpha_Ne_dict['1e+07']

fig,axes = plt.subplots(1,2,figsize=[11,5],facecolor='w')
color_list = ['r','b','g','y']

ax1, ax2 = axes

j = 0
for key_gene, key_env in key_pair_list:
    key = key_gene
    item = np.array(lambda_dict[key_gene][key_env])[[0,1,3,4]]
    ax1.bar(
        np.array([0,1,2,3])+j*0.15,
        np.log10(item),
        width=0.1,color=color_list[j],label=key
    )
    j+=1
ax1.set_xticks(np.array([0,1,2,3])+0.24)
ax1.set_xticklabels(
    ['Original','Significant','Original/10','Significant/10'],
    fontsize=13
)
ax1.set_ylabel(r"$\it{\lambda}$",rotation=0,fontsize=16)
ax1.set_yticks([0,1,2,3,4,5,6])
ax1.set_yticklabels(['$10^{0}$','$10^{1}$','$10^{2}$','$10^{3}$','$10^{4}$','$10^{5}$','$10^{6}$'])
ax1.spines.right.set_visible(False)
ax1.spines.top.set_visible(False)

#ax1.legend(loc=[0.63,0.75])

######################################################

j = 0
for key_gene, key_env in key_pair_list:
    key = key_gene
    item = np.array(alpha_dict[key_gene][key_env])[[0,1,3,4]]
    ax2.bar(
        np.array([0,1,2,3])+j*0.15,
        np.log10(np.array(item) + 1e-20),
        width=0.1,color=color_list[j],label=key
    )
    j+=1
#ax2.hlines(y=np.log10(1-0.1),xmin=-1,xmax=4,color='k',linestyles='--')
ax2.set_xticks(np.array([0,1,2,3])+0.24)
ax2.xaxis.set_ticks_position('top') 
ax2.set_xticklabels(
    ['Original','Significant','Original/10','Significant/10'],
    fontsize=13
)
ax2.set_ylabel(r"$1-\it{\alpha}$",rotation=0,fontsize=16)
ax2.legend(fontsize=14)
ax2.set_ylim(-12,0)
#ax2.set_xlim(-0.2,3.6)
ax2.set_yticks([-12,-10,-8,-6,-4,-2,0])
ax2.set_yticklabels(['$<10^{-12}$','$10^{-10}$','$10^{-8}$','$10^{-6}$','$10^{-4}$','$10^{-2}$','$10^{0}$'])
ax2.spines.right.set_visible(False)
ax2.spines.bottom.set_visible(False)

plt.tight_layout()
plt.savefig('./Figures/Fig1_lower_new.pdf')
```
