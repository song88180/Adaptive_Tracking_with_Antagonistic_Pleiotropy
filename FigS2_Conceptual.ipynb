```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy.random as nrand
from copy import copy
from matplotlib.patches import FancyBboxPatch
```

```python
fig, axes = plt.subplots(3,3,figsize=[15,6])
```

```python
rcParams_dict = {
    'xtick.bottom': False,
    'axes.spines.top': True,
    'axes.spines.bottom': True,
    'axes.spines.right': True,
    'axes.spines.left': True,
    'patch.edgecolor':'k',
    'patch.force_edgecolor': True,
    'font.size': 15,
    'pdf.fonttype':42,
    'font.family':'Arial',
    'mathtext.fontset':'custom',
    'mathtext.default': 'regular',
    'mathtext.it': 'Arial:italic',
    'mathtext.it': 'Arial:italic',
}
mpl.rcParams.update(rcParams_dict)
```

```python
rcParams_dict = {
    'xtick.bottom': False,
    'ytick.left': False,
    'xtick.labelbottom':False,
    'ytick.labelleft':False,
    'axes.spines.top': False,
    'axes.spines.bottom': False,
    'axes.spines.right': False,
    'axes.spines.left': False,
    'hatch.linewidth': 4,
    'patch.edgecolor':'k',
}
mpl.rcParams.update(rcParams_dict)
fig, axes = plt.subplots(3,1,figsize=[5,6])
ax1,ax2,ax3 = axes

width_list = [
    [0.9,0.1],
    [0.75,0.1,0.15],
    [0.875,0.1,0.025]
]

left_list = [
    [0,0.9],
    [0,0.75,0.85],
    [0,0.875,0.975]
]

for i,ax in enumerate(axes):
    ax.barh(
        y=0.5,width=width_list[i],left=left_list[i],
        color=['C10','C7','C3'],height=0.4,
        linewidth=1,edgecolor='k'
    )
    
    ax.set_xlim(0,1.1)
    ax.set_ylim(0,1)
    
    new_patches = []
    for patch in reversed(ax.patches):
        bb = patch.get_bbox()
        color=patch.get_facecolor()
        p_bbox = FancyBboxPatch((bb.xmin, bb.ymin),
                            abs(bb.width), abs(bb.height),
                            boxstyle="round,pad=-0.0040,rounding_size=0.01",
                            ec="k", fc=color,
                            mutation_aspect=4
                            )
        patch.remove()
        new_patches.append(p_bbox)
    for patch in new_patches:
        ax.add_patch(patch)
    
    if i == 1:
        bar = ax.patches[0]
        main_color = 'tab:blue'
        hatch_color = 'tab:red'
        rect = copy(bar)
        rect.set_facecolor('none')
        ax.add_patch(rect)
        bar.set_facecolor(main_color)
        bar.set_edgecolor(hatch_color)
        bar.set_hatch('//')
    
plt.savefig('./Conceptual2.pdf')
```
